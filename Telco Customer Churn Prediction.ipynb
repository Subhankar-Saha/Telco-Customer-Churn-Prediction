# Preprocessing Data

```python
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder
lencoder = LabelEncoder()
```

```python
for i  in range(0,len(df)):
    if(df['TotalCharges'][i] == ' '):
        df = df.drop([i])
```

```python
df['gender'] = pd.get_dummies(df['gender'],drop_first= True)
df['Churn'] = pd.get_dummies(df['Churn'],drop_first=True)
df['Partner'] = pd.get_dummies(df['Partner'],drop_first=True)
df['Dependents'] = pd.get_dummies(df['Dependents'],drop_first=True)
df['PhoneService'] = pd.get_dummies(df['PhoneService'],drop_first=True)
```

```python
df['OnlineSecurity'] = lencoder.fit_transform(df['OnlineSecurity'])
df['OnlineBackup'] = lencoder.fit_transform(df['OnlineBackup'])
df['DeviceProtection'] = lencoder.fit_transform(df['DeviceProtection'])
df['TechSupport'] = lencoder.fit_transform(df['TechSupport'])
df['StreamingTV'] = lencoder.fit_transform(df['StreamingTV'])
df['StreamingMovies'] = lencoder.fit_transform(df['StreamingMovies'])
df['Contract'] = lencoder.fit_transform(df['Contract'])
df['PaperlessBilling'] = lencoder.fit_transform(df['PaperlessBilling'])
df['PaymentMethod'] = lencoder.fit_transform(df['PaymentMethod'])
df['MultipleLines'] = lencoder.fit_transform(df['MultipleLines'])
df['InternetService'] = lencoder.fit_transform(df['InternetService'])
```

```python
df
```

```python
df['gender'] = df['gender'].astype('int64')
df['Partner'] = df['Partner'].astype('int64')
df['Dependents'] = df['Dependents'].astype('int64')
df['PhoneService'] = df['PhoneService'].astype('int64')
df['MultipleLines'] = df['MultipleLines'].astype('int64')
df['InternetService'] = df['InternetService'].astype('int64')
df['OnlineSecurity'] = df['OnlineSecurity'].astype('int64')
df['OnlineBackup'] = df['OnlineBackup'].astype('int64')
df['DeviceProtection'] = df['DeviceProtection'].astype('int64')
df['TechSupport'] = df['TechSupport'].astype('int64')
df['StreamingTV'] = df['StreamingTV'].astype('int64')
df['StreamingMovies'] = df['StreamingMovies'].astype('int64')
df['Contract'] = df['Contract'].astype('int64')
df['PaperlessBilling'] = df['PaperlessBilling'].astype('int64')
df['PaymentMethod'] = df['PaymentMethod'].astype('int64')
df['Churn'] = df['Churn'].astype('int64')
df['TotalCharges'] = df['TotalCharges'].astype('float64')
```

# Model Selection

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
clf = RandomForestClassifier()
```

```python
X = df.iloc[:,[0,1,2,3,6,7,8,9,10,11,13,14,15,16,17,18]]
y = df.iloc[:,-1]
```

```python
X
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42,shuffle = False)
y_train = y_train.reshape(-1,1)
y_test = y_test.reshape(-1,1)
```

```python
len(X_train)
```

```python
len(X_test)
```

```python
clf.fit(X_train,y_train)
```

```python
clf.feature_importances_
```

```python
pred = clf.predict(X_test)
```

```python
confusion_matrix(y_test, pred)
```

```python
accuracy_score(y_test, pred)
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression(n_jobs=4,random_state=0)
```

```python
lr.fit(X_train,y_train)
```

```python
pred2 = lr.predict(X_test)
```

```python
lr.score(X_test, y_test)
```

```python
confusion_matrix(y_test, pred2)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, pred2))
```
